## Two Files Split by Time

```python
# Load all files within 1 minute into a single a DataFrame
flows50_all = pd.concat([pd.read_csv('../data/50.flows', delimiter='|'), pd.read_csv('../data/50-1.flows', delimiter='|')], ignore_index=True)
flows51_all = pd.concat([pd.read_csv('../data/51.flows', delimiter='|'), pd.read_csv('../data/51-1.flows', delimiter='|')], ignore_index=True)
```

```python
# Sort them based on flow ending timestamp as we do in the program
flows50_sorted = flows50_all.sort_values(by='FLOW_END_MILLISECONDS', ignore_index=False, kind='mergesort')
flows51_sorted = flows51_all.sort_values(by='FLOW_END_MILLISECONDS', ignore_index=False, kind='mergesort')
```

### Visual Analysis

```python
# Print head and tail of the continuous unsorted files
flows50_all['FLOW_END_MILLISECONDS'].tail()
```

```python
flows51_all['FLOW_END_MILLISECONDS'].head()
```

```python
# Compute differences of first and last elements in the continuous files --> unsorted
(flows51_all.iloc[0]['FLOW_END_MILLISECONDS'] - flows50_all.iloc[len(flows50_all) - 1]['FLOW_END_MILLISECONDS']) / 1000
```

```python
# Print head and tail of the continuous sorted files
flows50_sorted['FLOW_END_MILLISECONDS'].tail()
```

```python
flows51_sorted['FLOW_END_MILLISECONDS'].head()
```

```python
# Compute differences of first and last elements in the continuous files --> sorted
(flows51_sorted.iloc[0]['FLOW_END_MILLISECONDS'] - flows50_sorted.iloc[len(flows50_sorted) - 1]['FLOW_END_MILLISECONDS']) / 1000
```
