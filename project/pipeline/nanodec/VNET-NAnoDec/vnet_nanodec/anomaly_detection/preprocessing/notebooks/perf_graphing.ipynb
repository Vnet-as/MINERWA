### Perf Graphing

```python
import numpy as np
import pandas as pd
```

```python
data_times = np.array([[0, 1, 2, 4, 6, 8, 10, 12, 14, 16], [385.901, 384.796, 204.406, 131.539, 106.503, 109.948, 110.538, 111.215, 113.495, 116.324]])
data_fps   = np.array([[0, 1, 2, 4, 6, 8, 10, 12, 14, 16], [2130, 2169, 4083, 6346, 7838, 7592, 7551, 7505, 7355, 7176]])

df_times = pd.DataFrame(data_times.T, columns=['Worker Processes', 'Processing Time']).astype({'Worker Processes': int})
df_fps   = pd.DataFrame(data_fps.T, columns=['Worker Processes', 'Flows per second'], dtype=int)
```

```python
df_times.plot.line(x='Worker Processes', y='Processing Time', title='Sample File Processing Time Based on Number of Worker Processes')
```

```python
df_fps.plot.line(x='Worker Processes', y='Flows per second', title='Flows Per Second Processing Based on Number of Worker Processes')
```
